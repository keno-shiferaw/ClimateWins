## 1. Import libraries and dataframe

```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import operator
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from matplotlib.pyplot import figure
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
path = r'C:\Users\kshifera\Desktop\CF_Course\Specialization\Project\Data Sets'
```

```python
#import dataframe
df = pd.read_csv(os.path.join(path, 'Task2_1.csv'))
```

```python
#check
df.head()
```

```python
#drop 'Unnamed: 0' column
df = df.drop(columns = ['Unnamed: 0'], axis =1)
df.head()
```

```python
#define cols
cols = df.columns
```

## 2. Run cluster analysis

### 3.1. Single Method

```python
dist_sin_SM = linkage(df.loc[:,
                        cols],method="single")
plt.figure(figsize=(18,6))
dendrogram(dist_sin_SM, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Single Method",fontsize=18)
plt.show()
```

### 3.2. Complete Method

```python
dist_sin_CM = linkage(df.loc[:,
                        cols],method="complete")
plt.figure(figsize=(18,6))
dendrogram(dist_sin_CM, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Complete Method",fontsize=18)
plt.show()
```

### 3.3. Average Method

```python
dist_sin_AM = linkage(df.loc[:,
                        cols],method="average")
plt.figure(figsize=(18,6))
dendrogram(dist_sin_AM, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Average Method",fontsize=18)
plt.show()
```

### 3.4. Ward Method

```python
dist_sin_WM = linkage(df.loc[:,
                        cols],method="ward")
plt.figure(figsize=(18,6))
dendrogram(dist_sin_WM, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Ward Method",fontsize=18)
plt.show()
```
